```python
# Cleaning data
# Fill NAN budget with 0
df['budget_musd'].fillna(0, inplace = True)

# Fill NaN votes with 0
df['vote_count'].fillna(0, inplace = True)
```

```python
# Let's create a column called profit
df['profit'] = df['revenue_musd'] - df['budget_musd']
df.head()
```

```python
# Create ROI column
# .loc[df_best.Budget >= 5]
df['roi'] = df['revenue_musd']/df['budget_musd'] 
```

```python
df.info()
```

```python
def top_n (column,n = 5,data = df,ascending = False,min_budget = 0,min_vote = 0):
     """
          This functions gets the data and the column
          filters your data set by a min budget and vote count
          sort the value by the information you want to know
          By default, it returns the top 5
          n: If you want to change the number
          if you want bottom, change 'ascending to False'
     """
     df2 = (
          data
          .loc[(data['budget_musd'] >= min_budget) & (data['vote_count'] >= min_vote),['title','poster_path', column]]
          .sort_values(by=[column], ascending=ascending)
          .head(n)
          .copy()
     )
     return HTML(df2.to_html(escape=False))
```

```python
df_clean = (df[[
                "poster_path"
                , "title"
                , "budget_musd"
                , "revenue_musd"
                , "vote_count"
                , "vote_average"
                , "popularity"
                , "profit"
                , "roi"]]
                .copy()
                )
```

```python
# Top 5 highest revenue 
top_n('revenue_musd',data=df_clean,n=5)
```

```python
# Creates a dataframe to aggregate metrics grouping by title
franchise = (
    df
    .groupby("belongs_to_collection")
    .agg({
        "title":"count",
        "budget_musd": ["sum","mean"], 
        "revenue_musd": ["sum","mean"], 
        "vote_average": "mean",
        "popularity": "mean", 
        "roi":"median", 
        "vote_count":"mean"})
)
```

```python
franchise
```

```python
# Top 5 franchise with biggest number os movies
franchise.nlargest(5,("title","count"))
```

```python
# Top 20 franchise order by mean revenue
franchise.nlargest(20, ("revenue_musd", "mean"))
```

```python
# Top 20 franchise order by mean budget
franchise.nlargest(20, ("budget_musd", "mean"))
```

```python
# Select the franchises with more than 1000 mean votes, 
# select top 20 and order by the vote_average

(
    franchise[franchise[("vote_count", "mean")] >=1000]
    .nlargest(20, ("vote_average", "mean"))
)
```

# Most Successful directors

```python
# Top 20 most commons directors
directors_top_20 = df['director'].value_counts().head(20)
directors_top_20
```

```python
plt.figure(figsize = (12, 8))
directors_top_20.plot(kind='bar', fontsize = 15)
plt.title("Most Active Directors",fontsize = 20)
plt.ylabel("Number of Movies", fontsize = 15)
plt.show()
```

```python
# Top 20 most profitable directors

top_20_profitable_directors = (
    df[['director','revenue_musd']]
    .groupby('director')
    .sum()
    .nlargest(20,'revenue_musd')
)

top_20_profitable_directors
```

```python
plt.figure(figsize = (12, 8))
top_20_profitable_directors.plot(kind='bar', fontsize = 15)
plt.title("Total Revenue",fontsize = 20)
plt.ylabel("Revenue (in MUSD)", fontsize = 15)
plt.show()
```
