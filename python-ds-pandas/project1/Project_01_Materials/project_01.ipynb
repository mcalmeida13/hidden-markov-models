```python
# Cleaning data
# Fill NAN budget with 0
df['budget_musd'].fillna(0, inplace = True)

# Fill NaN votes with 0
df['vote_count'].fillna(0, inplace = True)
```

```python
# Let's create a column called profit
df['profit'] = df['revenue_musd'] - df['budget_musd']
df.head()
```

```python
# Create ROI column
# .loc[df_best.Budget >= 5]
df['roi'] = df['revenue_musd']/df['budget_musd'] 
```

```python
df.info()
```

```python
def top_n (column,n = 5,data = df,ascending = False,min_budget = 0,min_vote = 0):
     """
          This functions gets the data and the column
          filters your data set by a min budget and vote count
          sort the value by the information you want to know
          By default, it returns the top 5
          n: If you want to change the number
          if you want bottom, change 'ascending to False'
     """
     df2 = (
          data
          .loc[(data['budget_musd'] >= min_budget) & (data['vote_count'] >= min_vote),['title','poster_path', column]]
          .sort_values(by=[column], ascending=ascending)
          .head(n)
          .copy()
     )
     return HTML(df2.to_html(escape=False))
```

```python
df_clean = (df[[
                "poster_path"
                , "title"
                , "budget_musd"
                , "revenue_musd"
                , "vote_count"
                , "vote_average"
                , "popularity"
                , "profit"
                , "roi"]]
                .copy()
                )
```

```python
# Top 5 highest revenue 
top_n('revenue_musd',data=df_clean,n=5)
```

```python
# Top 5 highest budget 
top_n('budget_musd',data=df_clean,n=5)
```

```python
# Top 5 highest profit 
top_n('profit',data=df_clean,n=5)
```

```python
# Top 5 lowest profit 
top_n('profit',data=df_clean,n=5,ascending = True)
```

```python
# Top 5 Highest ROI 
top_n('roi',data=df_clean,n=5,min_budget=50)
```

```python
# Top 5 lowest ROI 
top_n('roi',5,data=df_clean,ascending=True,min_budget = 100)
```

```python
# Search 1: Science Fiction Action Movie with Bruce Willis (sorted from high to low Rating)
scifi_action = df['genres'].str.contains("Science Fiction") & df['genres'].str.contains("Action")
bruce_williams = df['cast'].str.contains("Bruce Willis")

df_temp = df.loc[scifi_action & bruce_williams]
```

```python
top_n('vote_average',data=df_temp)
```

__Search 2: Movies with Uma Thurman and directed by Quentin Tarantino (sorted from short to long runtime)__

```python
tarantino = df['director'].str.contains("Quentin Tarantino")
uma_thurman = df['cast'].str.contains("Uma Thurman")

df_temp2 = df.loc[tarantino & uma_thurman]
top_n('runtime',data=df_temp2,ascending=True)
```

__Search 3: Most Successful Pixar Studio Movies between 2010 and 2015 (sorted from high to low Revenue)__

```python
year_interval = df.release_date.between("2010-01-01","2015-12-31")
pixar = df['production_companies'].str.contains("Pixar").fillna(False)

df_temp3 = df.loc[year_interval & pixar]
top_n('revenue_musd',data=df_temp3, n=40)
```

```python
scifi_action = df['genres'].str.contains("Thriller") | df['genres'].str.contains("Action")
english = df['original_language'] == 'en'
vote_75 = df['vote_average'] >= 7.5
vote_co = df.vote_count >= 10

df_temp4 = df.loc[scifi_action & english & vote_75 & vote_co]
top_n('release_date',data=df_temp4)
```
