```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = 'https://raw.githubusercontent.com/lazyprogrammer/machine_learning_examples/master/linear_regression_class/data_2d.csv'
df = pd.read_csv(path, names=['x_1','x_2','y'])
df.head()
```

```python
X = df.drop('y', axis=1)
Y = df['y']
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter(X['x_1'],X['x_2'],Y)
plt.show()
```

Y = AX

(X^T X)w = X^T y





```python
w = np.linalg.solve(np.dot(X.T,X),np.dot(X.T,Y))
```

```python
Yhat = np.dot(X,w)
```

```python
d1 = Y - Yhat
d2 = Y- Y.mean()
r2 = 1 -d1.dot(d1)/d2.dot(d2)
print("the r-squared is:", r2)
```
